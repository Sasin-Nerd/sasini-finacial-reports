```python
import  os
from IPython.display import  display
import  tabula
files  =  os.listdir()
dataframes   =  []
pdf_files = [file for file in files if file.lower().endswith('.pdf')]
for pdf in pdf_files:
  df = tabula.read_pdf(pdf, pages  = "all")
  dataframes.append(df)
```
